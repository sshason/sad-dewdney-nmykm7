```python
import json
import os

brawlstarts_old = json.load(open('src/assets/brawls.json'))
brawlstarts_new = json.load(open('src/assets/brawls_new.json'))

print('Old:', len(brawlstarts_old))
print('New:', len(brawlstarts_new))
```

```python
for key in brawlstarts_old:
    print(key)
```

```python
rarity = ['Trophy Road', 'Rare', 'Super Rare', 'Epic', 'Mythic', 'Legendary']
rarity
```

```python
all_brawlstars_old = set()
for category in brawlstarts_old:
    for old_key in category:
        all_brawlstars_old.add(old_key)

print('Old:', len(all_brawlstars_old))
```

```python
brawlstarts_new = set(brawlstarts_new)
```

```python
old_names = set([name.split('/')[-1].split('.png')[0] for name in all_brawlstars_old])
old_names
```

```python
new_names = set([name.split('/')[-1].split('.png')[0] for name in brawlstarts_new])
new_names
```

```python
map_old_new = {
    'Larry-&-Lawrie': 'larry_and_lawrie',
    'larry_and_lawrie': 'Larry-&-Lawrie',
    'El-Primo': 'el_primo',
    'el_primo': 'El-Primo',
    'R-T': 'r_t',
    'r_t': 'R-T',
    '8-Bit': '8_bit',
    '8_bit': '8-Bit',
    'Mr-P': 'mr_p',
    'mr_p': 'Mr-P',
}
```

```python
old_names_lower = set([name.lower() for name in old_names])
def map_name(name):
    if name.lower() in old_names_lower and name.lower() in new_names:
        return name
    if name in map_old_new:
        return map_old_new[name]
    else:
        print('Not found:', name)
        return name
```

```python
count = 0
for name in old_names:
    if map_name(name).lower() not in new_names:
        count += 1
        print(name)

for name in new_names:
    if map_name(name).lower() not in old_names_lower:
        count += 1
        print(name)

print(count)
```

```python
new_to_old_map = {map_name(name).lower(): name for name in old_names}
old_to_new_map = {v: k for k, v in new_to_old_map.items()}
```

```python
old_dict = {oldImageUrl.split('/')[-1].split('.png')[0]: oldImageUrl for oldImageUrl in all_brawlstars_old}
old_dict
```

```python
new_dict = {map_name(imageUrl.split('/')[-1].split('.png')[0]): imageUrl for imageUrl in brawlstarts_new}
new_dict
```

```python
new_data = {}

for index, category in enumerate(brawlstarts_old):
    category_name = rarity[index]
    for old_image_url in category:
        old_key = old_image_url.split('/')[-1].split('.png')[0]
        new_key = map_name(old_to_new_map[old_key])
        new_image_url = new_dict[new_key]
        new_data[old_key] = {
            'name': old_key,
            'category': category_name,
            'imageUrlNew': new_image_url,
            'imageUrlOld': old_image_url
        }

new_data
```

```python
json.dump(new_data, open('src/assets/brawls_new.json', 'w'), indent=2)
```
